**Tuples & Sets**

```python
pets = ("cat", "dog", "lion") #Tuples
pets = pets + ("rabbit",)
print (pets)
```

```python
cars = {"Toyota", "BMW", "Mercedes"} #sets
cars.add("Tesla")
print(cars)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
cars.remove("BMW")
print(cars)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
trucks = {"Volvo", "Scania", "Toyota"}

cars_but_not_trucks = cars.difference(trucks)
print(cars_but_not_trucks)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
trucks = {"Volvo", "Scania", "Toyota"}

trucks_but_not_cars = trucks.difference(cars)
print(trucks_but_not_cars)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
trucks = {"Volvo", "Scania", "Toyota"}

not_in_both = cars.symmetric_difference(trucks)
print(not_in_both)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
trucks = {"Volvo", "Scania", "Toyota"}

both_cars_and_trucks = cars.intersection(trucks)
print(both_cars_and_trucks)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
trucks = {"Volvo", "Scania", "Toyota"}

vehicles = cars.union(trucks)
print(vehicles)
```

```python
cars = {"Toyota", "BMW", "Mercedes"}
trucks = {"Volvo", "Scania", "Toyota"}

add_more_vehicles = cars.update(trucks)
print(cars)
```
